```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
        self.parent = None
    
    def add_child(self,child):
        self.children.append(child)
        child.parent = self
        
        return child
        
    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level
```

```python
root = TreeNode("Electronics")
```

```python
root.children.clear()
```

```python
for cat in ["Laptop", "Cellphone", "TV"]:
    root.add_child(TreeNode(cat))
```
